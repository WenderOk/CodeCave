```python
def funcX(x):
    if x <=- 2: return(1-(x+2)**2)
    elif -2 < x <= 2: return(-(x/2))
    elif 2 < x: return((x-2)**2+1)
funcX(float(input()))
```

# Этрусский перевод
---

```python
def ArabToRoman(num):
    nums = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
    romnums = ["M", "CM", "D", "CD", "C", "XC", "L", "XL", "X", "IX", "V", "IV", "I"]
    result = ''
    for i in range(len(nums)):
        while num >= nums[i]:
            result += romnums[i]; num -= nums[i]
    return result
ArabToRoman(int(input()))
```

# Шаш и Златопольский
---

```python
def findLastLetter(sen1,sen2, letter):
    print(f'Предложение 1: {sen1}\nПредложение 2: {sen2}')
    if sen2.rfind(letter) > sen1.rfind(letter): 
        print("Предложение 2, символ номер:")
        return(sen2.rfind(letter))
    else: 
        print("Предложение 1, символ номер:")
        return(sen1.rfind(letter))
findLastLetter(input("Введите первое предложение: "), input("Введите второе предложение: "), input("Введите букву, которую нужно найти: "))
        
```

# Перестановка элементов

```python
def swap(A,m):
    print(A)
    A[:] = A[m:] + A[:m]
    return(A)
swap([int(x) for x in input().split()], int(input()))
```

# Мой переезд из Европы
---

```python
def MilesToKm(n,KmToM):
    if KmToM: return(round(n/1.60934, 5))
    else: return(round(n*1.60934, 5))
MilesToKm(int(input()), bool(input()))
```

# Число восьмых
---

```python
def CountEighths(N):
    num = int(N[:N.find('/')])
    den = int(N[N.find('/')+1:])
    return(int(8/den*num))
CountEighths(input())
```

# Футбольное поле
---

```python
# Да, я использовал 2д массив, но без него тут с этим "замечательном" вводом координат пришлось бы туго. 
# Ну и это ведь всего лишь массив в массиве - ничего координально нового
def FootballField(n, corn, coords):
    counter=0
    for i in range(n):
        if (float(corn[0][0]) < float(coords[i][0]) < float(corn[1][0])) and (float(corn[0][1]) < float(coords[i][1]) < float(corn[1][1])): 
            counter += 1
    return(counter)
N = int(input())
FootballField(N, [input().split() for _ in range(2)], [input().split() for _ in range(N)])
```
