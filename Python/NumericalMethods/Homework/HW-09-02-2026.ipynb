```python
# Basics of symbolyc expressions
from sympy import symbols

x,y,z = symbols('x y z')
expr = x**2 + y**2 + z**2
print(type(expr))
expr # jupyter имба
```

```python
# Opening of brackets
from sympy import expand

x,y = symbols('x y')
expr = (x + y)**3 * (x - y)**2
print(expand(expr))
expand(expr)
```

```python
# Factorization
from sympy import factor

x,y = symbols('x y')
expr = x**4 - 5*x**3 + 5*x**2 - 5*x + 4
print(factor(expr))
factor(expr)
```

```python
# Simplifying and substitution
from sympy import simplify

x = symbols('x')
expr = (x**2 - 4) / (x - 2)
expr = simplify(expr)
print(f"Simplified: {expr}")
print(f"Result: {expr.subs(x, 5)}")
```

```python
# Combinated usage

A, B = symbols("A B")
S = A**2 * B + 2 * A * B**2 + B**3
H = A - B
V = S * H
V_exp = expand(V)
print(f"Expamded: {V_exp}")
print(f"Factored: {factor(V_exp)}")
print(f"Factored original: {factor(V)}")
```
