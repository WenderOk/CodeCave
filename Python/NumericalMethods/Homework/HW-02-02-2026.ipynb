```python
# Распределение чаевых
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
plt.figure(figsize=(10, 6))
sns.histplot(data=tips, x='tip', kde=True)
plt.title('Tip distribution')
plt.xlabel('Sum of tips ($)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Связь между общим счётом и чаевыми
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', 
                hue='smoker',
                style='smoker',
                s=100,
                alpha=0.7,
                palette=['red', 'green'])
plt.xlabel('Total bill ($)', fontsize=12)
plt.ylabel('Tips ($)', fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--')
sns.regplot(data=tips[tips['smoker'] == 'Yes'], x='total_bill', y='tip', 
            scatter=False, color='red', label='Smoking (trend)', ci=None)
sns.regplot(data=tips[tips['smoker'] == 'No'], x='total_bill', y='tip', 
            scatter=False, color='green', label='Not smoking (trend)', ci=None)
plt.legend(title='Smoking customer', fontsize=11, title_fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Средний счёт по дням недели
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

plt.figure(figsize=(10, 6))
sns.barplot(data=tips, x='day', y='total_bill')

plt.title('Average total bill by day of week')
plt.xlabel('Day of week ')
plt.ylabel('Average bill ($)')

ax = plt.gca()
for bar in ax.patches:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.5,
            f'${height:.2f}', ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Сравнение распределения счёта по дням
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

plt.figure(figsize=(10, 6))
sns.boxplot(data=tips, x='day', y='total_bill')

plt.title('Distribution of total bill by day of week')
plt.xlabel('Day of week')
plt.ylabel('Total bill ($)')

plt.show()
```
