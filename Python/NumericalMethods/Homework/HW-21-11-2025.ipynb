```python
with open("data/24-001.txt", "r") as F:
    max_len, amount = 1, 1
    S = F.read()
    for i in range(len(S)-1):
        if S[i] == S[i+1]: 
            amount+=1
            max_len = max(amount, max_len)
        else: amount = 1
    print(max_len)
```

```python
with open("data/24.txt", "r", encoding="UTF-8") as F:
    max_len, amount, symbol_idx = 1, 0, 0 
    S = F.read()
    for i in range(len(S)-1):
        if S[symbol_idx] != S[i+1]: 
            amount+=1
            max_len = max(amount, max_len)
        else: 
            amount = 0
            symbol_idx = i+1        
    print(max_len)
```

```python
with open("data/k7a-5.txt", "r", encoding="UTF-8") as F:
    max_len, amount = 1, 1
    S = F.read()
    for i in range(len(S)-1):
        if S[i] in "CF": amount = 0
        else: 
            amount+=1
            max_len = max(amount, max_len)
    print(max_len)
```

```python
with open("data/k7c-1.txt", "a", encoding="UTF-8") as F:
    S = F.readline()
    amount = 0
    for i in range(len(S)-2):
        if (S[i] in "BCD" and
        S[i+1] in "BDE" and S[i+1] != S[i] and
        S[i+2] in "BCE" and S[i+2] != S[i+1]):
            amount += 1
    print(amount)
```

```python
result, min_s = "", 1
with open("data/input.txt","r") as F:
    for s in F:
        s = s.strip()
        result += f"{min_s}) {s} {len(s)}\n"
        min_s+=1
with open("data/input.txt", "w") as F:
    F.write(result)
```

```python
short_lines, min_len = [], float('inf')
with open("data/input2.txt", 'r') as F:
    for l in F:
        cur_line = l.strip()
        if len(cur_line) < min_len:
            min_len = len(cur_line)
            short_lines = [cur_line]
        elif len(cur_line) == min_len:
            short_lines.append(cur_line)

with open("data/output.txt", 'w') as F:
    for l in short_lines:
        F.write(l + '\n')
```
