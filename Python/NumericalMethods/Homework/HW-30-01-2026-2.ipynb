```python
# 4
import numpy as np

def monte_carlo_area(r, n=1000000):
    points = np.random.uniform(-r, r, (n, 2))
    distances = np.sqrt(points[:, 0]**2 + points[:, 1]**2)
    inside = distances <= r
    return (2*r)**2 * np.sum(inside) / n

print("R   Exact   Monte-Carlo  Coefficient")
circles = [1, 2, 3, 4, 5]

for r in circles:
    exact = np.pi * r**2
    mc = monte_carlo_area(r)
    coeff = mc / r**2
    print(f"{r}   {exact:.4f}   {mc:.4f}       {coeff:.4f}")
```

```python
import numpy as np
print("a) Zero vector with lenght 10:")
zero_vector = np.zeros(10)
print(
f"""    Form: {zero_vector.shape}
    Type: {zero_vector.dtype}
    Result: {zero_vector}""")

print("\nb) 3×3 matrix with zeros:")
zero_matrix = np.zeros((3, 3))
print(f"""   Form: {zero_matrix.shape}
   Type: {zero_matrix.dtype}
   Result:\n{zero_matrix}""")

print("\nc) Vector with values from 1 to 10:")
vector_1_to_10 = np.arange(1, 11)
print(f"""   Form: {vector_1_to_10.shape}
   Type: {vector_1_to_10.dtype}
   Result: {vector_1_to_10}""")

print("\nd) Reverse vector (10 9 8 ... 0), multiplied by 2:")
reverse_vector = np.arange(10, -1, -1)
reverse_vector_times_2 = reverse_vector * 2
print(f"""   Original vector (10..0): {reverse_vector}
   Multiplied by 2: {reverse_vector_times_2}
   Form: {reverse_vector_times_2.shape}""")

print("\ne) 2-dimensional 10x10 array with ones at edges and zeros in center:")
array_10x10 = np.zeros((10, 10), dtype=int)
array_10x10[0, :] = 1
array_10x10[-1, :] = 1
array_10x10[:, 0] = 1
array_10x10[:, -1] = 1

print(f"""   Form: {array_10x10.shape}
   Type: {array_10x10.dtype}
   Result:""")
print(array_10x10)
```

```python
import numpy as np
import matplotlib.pyplot as plt

sigmoid = lambda x: 1 / (1 + np.exp(-x))
x = np.linspace(-10, 10, 1000)

plt.figure(figsize=(8, 6))
plt.plot(x, sigmoid(x), 'b-', linewidth=3)
plt.xlabel('x')
plt.ylabel('σ(x)')
plt.grid(True)
plt.show()
```
