```python
# Inaccuracy 1
import decimal
x = 1 / 3
x_star = 0.333
abs_inaccur = abs(x - x_star)
rel_inaccur = abs_inaccur / abs(x)
diff = abs(x - x_star)
diff_str = format(diff, '.20f')
decimal.getcontext().prec = 30

print(f"""Absolute inaccuracy: {abs_inaccur}
Relative inaccuracy: {rel_inaccur}
Difference (accurate): {diff}
Difference (formatted): {diff_str}""")
print("\nDefining correct digits in wide meaning:")
m = -3
if abs_inaccur <= 0.5 * 10**m:
    print(f"All digits before {abs(m)} digit after point are correct.")
else:
    print("Not all digits are correct.")

print(f"""x = {x}
x* = {x_star}
Absolute inaccuracy = {abs_inaccur:.2e}
Relative inaccuracy = {rel_inaccur:.2e}""")
```

```python
# Inaccuracy 2
import fractions

x = fractions.Fraction(1, 3)
y = fractions.Fraction(7, 18)

x_star = 0.333
y_star = 0.389

delta_x = abs(float(x) - x_star)
delta_y = abs(float(y) - y_star)

rel_x = delta_x / abs(float(x))
rel_y = delta_y / abs(float(y))

print(f"""x = {x} ≈ {float(x):.10f}
y = {y} ≈ {float(y):.10f}
x* = {x_star}
y* = {y_star}
Δx = {delta_x:.2e}
Δy = {delta_y:.2e}
δx = {rel_x:.2e}
δy = {rel_y:.2e}\n""")

S = x + y
S_star = x_star + y_star
delta_S = delta_x + delta_y
print(f"""Accurate sum S = {S} ≈ {float(S):.10f}
Approximate sum S* = {S_star:.3f}
Estimation ΔS ≤ Δx + Δy = {delta_S:.2e}
Real ΔS_факт = {abs(float(S) - S_star):.2e}\n""")

D = x - y
D_star = x_star - y_star
delta_D = delta_x + delta_y
print(f"""Accurate difference D = {D} ≈ {float(D):.10f}
Approximate difference D* = {D_star:.3f}
Estimation ΔD ≤ Δx + Δy = {delta_D:.2e}
Real ΔD_факт = {abs(float(D) - D_star):.2e}\n""")

P = x * y
P_star = x_star * y_star
rel_P = rel_x + rel_y
delta_P = float(abs(P)) * rel_P
print(f"""Accurate product P = {P} ≈ {float(P):.10f}
Approximate product P* = {P_star:.5f}
Estimation δP ≈ δx + δy = {rel_P:.2e}
Estimation ΔP ≈ |P| * δP = {delta_P:.2e}
Real δP_факт = {abs(float(P) - P_star)/abs(float(P)):.2e}
Real ΔP_факт = {abs(float(P) - P_star):.2e}\n""")

Q = x / y
Q_star = x_star / y_star
rel_Q = rel_x + rel_y
delta_Q = float(abs(Q)) * rel_Q
print(f"""Accurate quotient Q = {Q} ≈ {float(Q):.10f}
Approximate quotient Q* = {Q_star:.5f}
Estimation δQ ≈ δx + δy = {rel_Q:.2e}
Estimation ΔQ ≈ |Q| * δQ = {delta_Q:.2e}
Real δQ_факт = {abs(float(Q) - Q_star)/abs(float(Q)):.2e}")
Real ΔQ_факт = {abs(float(Q) - Q_star):.2e}""")
```

```python
# Distance between 2 points on the Earth
from decimal import Decimal
import math

def earth_distance(lat1, lon1, lat2, lon2):
    R = Decimal('6371.01')
    lat1_r = math.radians(lat1)
    lon1_r = math.radians(lon1)
    lat2_r = math.radians(lat2)
    lon2_r = math.radians(lon2)
    angle = math.acos(
        math.sin(lat1_r) * math.sin(lat2_r) + 
        math.cos(lat1_r) * math.cos(lat2_r) * 
        math.cos(lon1_r - lon2_r)
    )
    return R * Decimal(angle)
```

```python
dist = earth_distance(55, 37, 59, 30)
print(f"Distance 1: {dist.quantize(Decimal('0.01'))} km")
print(f"Expected: 613.83 км")
```
