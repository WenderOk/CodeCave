```python
from functools import reduce

def char_count(text):
    chars = list(text)
    return reduce(lambda acc, char: {**acc, char: acc.get(char, 0) + 1}, chars, {})

text = input("Enter a string: ")
print(char_count(text))  
```

```python
humans = [
    {'name': 'Sergey', 'age': 47},
    {'name': 'Alibaba', 'age': 54},
    {'name': 'Jessy', 'age': 24},
    {'name': 'Walter', 'age': 55}
]
humans_sorted = sorted(humans, key=lambda x: (-x['age'], x['name']))

for human in humans_sorted:
    print(f"{human['name']}: {human['age']}")
```
