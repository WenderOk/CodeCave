```python
# 4
import matplotlib.pyplot as plt
import numpy as np

def create_square(size=15, num_points=7500000, point_size=15, alpha=0.5):
    x = np.random.uniform(-size/2, size/2, num_points)
    y = np.random.uniform(-size/2, size/2, num_points)
    colors = np.random.rand(num_points, 3)
    fig, ax = plt.subplots(figsize=(8, 8))
    scatter = ax.scatter(x, y, 
                        c=colors, 
                        s=point_size, 
                        alpha=alpha, 
                        edgecolors='none')
    plt.tight_layout()
    return fig, ax

fig2, ax2 = create_square(size=12, num_points=800000, point_size=5, alpha=0.5)

plt.show()
```

```python
# 4
import matplotlib.pyplot as plt
import numpy as np
print("Enter coords (x1 y1 x2 y2):")
x1, y1, x2, y2 = list(map(float, input().split()))
# print("Enter coords to exclude (x1 y1 x2 y2):")
# x1_ex, y1_ex, x2_ex, y2_ex = list(map(float, input().split()))
x_min = min(x1, x2)
x_max = max(x1, x2)
y_min = min(y1, y2)
y_max = max(y1, y2)
num_points = 100000
point_size = 15
alpha = 0.5
x = np.random.uniform(x_min, x_max, num_points)
y = np.random.uniform(y_min, y_max, num_points)
colors = np.random.rand(num_points, 3)
fig, ax = plt.subplots(figsize=(8, 6))
# if x.all() in range(int(min(x1_ex, x2_ex)), int(max(x1_ex, x2_ex))):
ax.scatter(x, y, c=colors, s=point_size, alpha=alpha, edgecolors='none')
width = x_max - x_min
height = y_max - y_min
rect = plt.Rectangle((x_min, y_min), width, height,
                    linewidth=2, edgecolor='black', facecolor='none')
ax.add_patch(rect)
ax.set_xlim(x_min - width*0.1, x_max + width*0.1)
ax.set_ylim(y_min - height*0.1, y_max + height*0.1)
ax.set_title(f'({x_min},{y_min})-({x_max},{y_max})')

plt.tight_layout()
plt.show()
```

```python
# 5
import matplotlib.pyplot as plt
import numpy as np
import math

def fill_triangle(base=10.0, points_num=100000):
    height = (base/2) * math.tan(math.radians(22.5))
    A = (0, 0)
    B = (base, 0)
    C = (base/2, height)
    r1 = np.random.rand(points_num)
    r2 = np.random.rand(points_num)
    sqrt_r1 = np.sqrt(r1)
    x = (1 - sqrt_r1) * A[0] + sqrt_r1 * (1 - r2) * B[0] + sqrt_r1 * r2 * C[0]
    y = (1 - sqrt_r1) * A[1] + sqrt_r1 * (1 - r2) * B[1] + sqrt_r1 * r2 * C[1]
    colors = np.random.rand(points_num, 3)
    plt.figure(figsize=(8, 8))
    plt.scatter(x, y, c=colors, s=10, alpha=0.7, edgecolors='none')
    triangle_x = [A[0], B[0], C[0], A[0]]
    triangle_y = [A[1], B[1], C[1], A[1]]
    plt.plot(triangle_x, triangle_y, 'k-', linewidth=2)
    
    plt.show()

points = int(input("Number of points: "))
base = float(input("Lenght of triangle base: "))
fill_triangle(base, points)
```
