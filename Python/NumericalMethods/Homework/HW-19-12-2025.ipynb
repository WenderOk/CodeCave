```python
# Домик
import tkinter as tk
from tkinter import simpledialog

def draw_building(x, y, size=150):
    canvas.delete("all")
    canvas.create_rectangle(x, y, x+size, y+size,
                            fill='green', outline='black')
    canvas.create_polygon(x-size//4, y, 
                         x+size//2, y-size//2, 
                         x+size+size//4, y, 
                         fill='orange', outline='black')
    win_width = size // 2.5
    win_height = size // 2
    win_x = x + (size - win_width) // 2
    win_y = y + (size - win_height) // 3.5
    canvas.create_rectangle(win_x, win_y, 
                           win_x + win_width, win_y + win_height, 
                           fill='black', outline='white', width=3)

def update_coords():
    x = simpledialog.askinteger("X coordinate", "Enter X coordinate:", parent=root)
    y = simpledialog.askinteger("Y coordinate", "Enter Y coordinate:", parent=root)
    if x and y: draw_building(x, y)

width = 500
height = 500
root = tk.Tk()
canvas = tk.Canvas(root, width=width, height=height, bg='blue')
canvas.pack()
btn = tk.Button(root, text="Change coordinates", command=update_coords)
btn.pack()

draw_building(width//2.8, height//2.3)
root.mainloop()
```

```python
# Лягушка
import tkinter as tk
from tkinter import simpledialog

def draw_frog(x, y):
    canvas.delete("all")
    mouth_w = 100
    mouth_h = 60
    eye_rad = 20

    upper_points = [
        x - mouth_w, y,
        x + mouth_w, y,
        x, y - mouth_h
    ]
    lower_points = [
        x - mouth_w, y,
        x + mouth_w, y,
        x, y + mouth_h
    ]
    canvas.create_polygon(upper_points, fill='lime', outline='black', width=2)
    canvas.create_polygon(lower_points, fill='green', outline='black', width=2)
    canvas.create_oval(
        x - mouth_w - eye_rad, 
        y - eye_rad,
        x - mouth_w + eye_rad, 
        y + eye_rad,
        fill='black', outline='black', width=2
    )
    canvas.create_oval(
        x + mouth_w - eye_rad, 
        y - eye_rad,
        x + mouth_w + eye_rad, 
        y + eye_rad,
        fill='black', outline='black', width=2
    )

def update_coords():
    x = simpledialog.askinteger("X coordinate", "Enter X coordinate:", parent=root)
    y = simpledialog.askinteger("Y coordinate", "Enter Y coordinate:", parent=root)
    if x and y: draw_frog(x, y)

width = 500
height = 500
root = tk.Tk()
canvas = tk.Canvas(root, width=width, height=height, bg='blue')
canvas.pack()
btn = tk.Button(root, text="Change coordinates", command=update_coords)
btn.pack()

draw_frog(width//2, height//2)
root.mainloop()
```

```python
# Корона
import tkinter as tk
from tkinter import simpledialog

def draw_crown(x,y):
    canvas.delete("all")
    crown_w = 100
    triangle_w = 150
    crown_height = 100
    triangle_points1 = [
        x + triangle_w // 2, y + crown_height,
        x - triangle_w // 2, y + crown_height,
        x, y - crown_height,
    ]
    triangle_points2 = [
        x + triangle_w // 2 - crown_w // 2, y + crown_height,
        x - triangle_w // 2 - crown_w // 2, y + crown_height,
        x - crown_w * 2, y - crown_height + 20,
    ]
    triangle_points3 = [
        x + triangle_w // 2 + crown_w // 2, y + crown_height,
        x - triangle_w // 2 + crown_w // 2, y + crown_height,
        x + crown_w * 2, y - crown_height + 20,
    ]
    canvas.create_polygon(triangle_points2, fill='yellow', outline='black', width=2)
    canvas.create_polygon(triangle_points3, fill='orange', outline='black', width=2)
    canvas.create_polygon(triangle_points1, fill='gold', outline='black', width=2)
    circle_radius = 12
    canvas.create_oval(
        x + crown_w * 2 - circle_radius,
        y - crown_height + 20 - circle_radius,
        x + crown_w * 2 + circle_radius,
        y - crown_height + 20 + circle_radius,
        fill='orange', outline='black', width=2
    )
    canvas.create_oval(
        x - crown_w * 2 - circle_radius,
        y - crown_height + 20 - circle_radius,
        x - crown_w * 2 + circle_radius,
        y - crown_height + 20 + circle_radius,
        fill='yellow', outline='black', width=2
    )
    canvas.create_oval(
        x - circle_radius,
        y - crown_height - circle_radius,
        x + circle_radius,
        y - crown_height + circle_radius,
        fill='gold', outline='black', width=2
    )
    

def update_coords():
    x = simpledialog.askinteger("X coordinate", "Enter X coordinate:", parent=root)
    y = simpledialog.askinteger("Y coordinate", "Enter Y coordinate:", parent=root)
    if x and y: draw_crown(x, y)

width, height = 1000, 1000
root = tk.Tk()
canvas = tk.Canvas(root, width=width, height=height, bg='blue')
canvas.pack()
btn = tk.Button(root, text="Change coordinates", command=update_coords)
btn.pack()

draw_crown(width//2, height//2)
root.mainloop()
```

```python
# 3
import tkinter as tk

def animate_square():
    global square_id, x_pos
    canvas.move(square_id, -5, 0)
    x_pos -= 5
    if x_pos < -50:
        canvas.coords(square_id, 900, 450, 850, 400)
        x_pos = 900
    canvas.after(10, animate_square)
root = tk.Tk()
canvas = tk.Canvas(root, width=900, height=900)
canvas.pack()
x_pos, square_size = 900, 50

square_id = canvas.create_rectangle(
    x_pos, 450,
    x_pos + square_size, 400,
    fill='yellow',
    outline='black',
    width=2
)
animate_square()
root.mainloop()
```

```python
# 4
import tkinter as tk

def animate_squares():
    global square1_id, square2_id, x1_pos, x2_pos
    canvas.move(square1_id, 5, 0)
    x1_pos += 5
    canvas.move(square2_id, -5, 0)
    x2_pos -= 5
    if x1_pos > 850:
        canvas.coords(square1_id, -50, 200, 0, 250)
        x1_pos = -50
    if x2_pos < -50:
        canvas.coords(square2_id, 800, 350, 850, 400)
        x2_pos = 800
    
    root.after(10, animate_squares)
root = tk.Tk()
canvas = tk.Canvas(root, width=800, height=600)
canvas.pack()
square_size, x1_pos = 50, -50
square1_id = canvas.create_rectangle(
    x1_pos, 200,
    x1_pos + square_size, 250,
    fill='yellow',
    outline='black',
    width=2,
    tags='square1'
)

x2_pos = 800
square2_id = canvas.create_rectangle(
    x2_pos, 350,
    x2_pos + square_size, 400,
    fill='yellow',
    outline='black',
    width=2,
    tags='square2'
)
animate_squares()
root.mainloop()
```

```python
# 4
import tkinter as tk

def animate_squares():
    global square1_id, square2_id, x1_pos, x2_pos, speed1, speed2
    canvas.move(square1_id, speed1, 0)
    x1_pos += speed1
    canvas.move(square2_id, speed2, 0)
    x2_pos += speed2
    if (x1_pos > 850) or (x1_pos < 0):
        speed1 = -speed1
    if (x2_pos > 850) or (x2_pos < 0):
        speed2 = -speed2
    
    root.after(20, animate_squares)
root = tk.Tk()
canvas = tk.Canvas(root, width=900, height=900, bg="blue")
canvas.pack()
square_size = 50
speed1, speed2 = 10, -10

x1_pos = -1
square1_id = canvas.create_rectangle(
    x1_pos, 200,
    x1_pos + square_size, 250,
    fill='yellow',
    outline='black',
    width=2,
    tags='square1'
)

x2_pos = 800
square2_id = canvas.create_rectangle(
    x2_pos, 350,
    x2_pos + square_size, 400,
    fill='yellow',
    outline='black',
    width=2,
    tags='square2'
)
animate_squares()
root.mainloop()
```
