```python
k=0
with open("input.txt","r") as F:
    n = int(F.readline())
    for _ in range(n):
        num = int(F.readline())
        if (num % 3 == 0) and (abs(num) % 10 == 2): k+=1
with open("output.txt", "w") as F:
    F.write(str(k))
```

```python
with open("input.txt","r") as F:
    s = F.readline()
max_len, cur_len = 0, 0
for i in range(1, len(s)-1):
    if (s[i] != s[i-1]) and (s[i] != s[i+1]): 
        cur_len+=1
        max_len = max(cur_len, max_len)
    else: cur_len = 1
print(max_len)
```

```python
max_len, cur_len = 0, 0
max_file  = ""
for i in range(101):
    with open(f"k7 0-100 data/k7-{i}.txt","r") as F:
        cur_len = 0
        for c in F.readline():
            if c == 'C':
                cur_len+=1
                if max_len < cur_len:
                    max_len = cur_len
                    max_file = f"k7-{i}.txt"
            else:
                cur_len = 0

print(max_len, max_file)
```

```python
with open("input.txt","r") as F: 
    s = F.readline()
max_len, cur_len, E = 0, 0, "EAB"
for i in range(1, len(s)-1):
    if s[i] == E[cur_len % 3]: 
        cur_len+=1
        max_len = max(cur_len, max_len)
    else: cur_len = 1
print(max_len)
```
