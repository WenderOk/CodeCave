## Зоопарк

```python
F = open("Зоопарк.txt", "r", encoding="UTF-8")
S = F.read().split()
S
```

```python
D = {}
for animal in S:
    D[animal] = D.get(animal, 0)+1
```

```python
with open("report.txt", "a", encoding="UTF-8") as F:
    Animals = sorted(D, key=D.get)
    R = "Rarest - " + str(Animals[0]) + "\n"
    for animal in Animals[1:]:
        R += f"{animal}: {D[animal]}\n"
    F.write(R)

F.close()
```

## Разбиение списка

```python
A = ["k1", 1, 2, 3, "k2", 1, 3, "k3", 1]
def func(arr):
    D = {}
    key = ""
    for item in arr:
        if type(item) == str:
            D[item] = []
            key = item
        else:
            D[key].append(item)
    print(D)

func(A)
```
