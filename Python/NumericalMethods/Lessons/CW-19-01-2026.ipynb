# Matplotlib в Python

Matplotlib используется для рисования графиков и диаграмм

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.plot([1,2,3,4],[65,6,44,43])
plt.xlabel("Ages")
plt.ylabel("Salary")
plt.title("Median Salary (USD) by Age")
plt.legend(["All devs", "Python"])
plt.show()
```

```python
# Сохраняет график
plt.savefig("plot.png")
plt.show()
```

```python
import numpy as np
# plt.style.use("fivethertyeight")
ages_x = list(range(25,36))
x_indexes = np.arange(len(ages_x))
width = 0.25
plt.bar(x_indexes - width, [1,4,9,16])
plt.bar(x_indexes + width, [65,6,44,43])
```

```python
import csv; from collections import Counter
import numpy as np
from matplotlib import pyplot as plt
with open("data/stackoverflow-dev-servey-2019.csv") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    language_counter = Counter()
    # row = next(csv_reader)
    # print(*row["LanguagesWorkedWith"].split(";"))
    for row in csv_reader:
        language_counter.update(row["LanguagesWorkedWith"].split(";"))
print(language_counter.most_common(15))
```

```python
languages = []
popularity = []
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
languages.reverse()
popularity.reverse()
```

```python
plt.barh(languages, popularity)
plt.title("Most Popular Languages")
plt.xlabel("Users")
plt.show()
```

# Круговые диаграммы

```python
slices = [60, 40]
plt.pie(slices)

plt.title("Круговые диаграммы")
plt.show()
```

```python
slices = [59219, 55266, 47544, 36443, 35917]
labels = ["JS", "HTML", "SQL", "Python", "Java"]
explode = [0,0,0,0.1,0]

plt.pie(slices, labels=labels, explode=explode, 
autopct="%1.1f%%",
shadow=True)
wedgeprops={"edgecolor": "black"}
plt.show()
```
