# Дискретизация

```python
def Area(f, a, b, n):
    h = (b - a) / n
    S = 0
    for i in range(1, n + 1):
        x_i = a + i * h
        S += f(x_i)
    return h * S

    
```

```python
def func(x): return 1 / x

a = 1.0
b = 2.0
n = 4

approx_integral = Area(func, a, b, n)
print(f"Приближенное значение интеграла n = {n}: {approx_integral}")

import math
exact_value = math.log(2)
print(f"{abs(exact_value - approx_integral)}")
```

```python
import tkinter, random, time
x, y, r, vel_x, vel_y = 100, 100, 50, 5, 5
W, H = 1280, 720
N = 100
x, y = [0] * N, [0] * N
vel_x, vel_y = [0] * N, [0] * N
time_old = time.time()

root = tkinter.Tk()
root.configure()
canvas = tkinter.Canvas(root, width=W, height=H)
canvas.pack()

def animate():
    global time_old, W, H, x, y, r, N, vel_x, vel_y
    time_now = time.time()
    dt = time_now - time_old
    time_old = time_now
    x += vel_x * dt
    y += vel_y * dt
    canvas.delete("all")
    canvas.create_oval(x - r, y - r, x + r, y + r, fill="red", outline="black")
    canvas.after(50, animate)

animate()
root.mainloop()
```

```python
import tkinter, random
x, y, r, vel_x, vel_y = 100, 100, 50, 5, 5
W, H = 1280, 720
N = 100
x, y = [0] * N, [0] * N
vel_x, vel_y = [0] * N, [0] * N

for i in range(N):
    x[i] = random.randint(r, W - r)
    y[i] = random.randint(r, H - r)
    vel_x[i] = 10
    vel_y[i] = 6


root = tkinter.Tk()
root.configure()
canvas = tkinter.Canvas(root, width=W, height=H)
canvas.pack()

def animate():
    global W, H, x, y, r, N, vel_x, vel_y
    canvas.delete("all")
    for i in range(N):
        if x[i] < r or x[i] > W - r:
            vel_x[i] *= -1.5
        if y[i] < r or y[i] > H - r:
            vel_y[i] *= -1.5
        x[i] += vel_x[i]
        y[i] += vel_y[i]
        canvas.create_oval(x[i] - r, y[i] - r, x[i] + r, y[i] + r, fill="red", outline="black")
    canvas.after(50, animate)

animate()
root.mainloop()
```
